```
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

temp = a
a = b
b = temp

print("After swapping: a =", a, ", b =", b)
```

```
import math

for i in range(1, 6):
    print("Factorial of", i, "=", math.factorial(i))
```

```
n = int(input("Enter how many terms: "))

a, b = 0, 1

print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```
